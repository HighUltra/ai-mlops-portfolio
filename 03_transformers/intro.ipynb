```python
from transformers import pipeline

# 1. Cargamos una "tubería" (pipeline) de análisis de sentimiento
# Esto descarga automáticamente un modelo pre-entrenado la primera vez
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

# 2. Probamos con un par de frases
results = classifier([
    "I am very happy with this machine learning project!",
    "I am frustrated because the code had an error."
])

for result in results:
    print(f"Texto: {result['label']}, Score: {round(result['score'], 4)}")
```

```python
import time

start = time.time()
classifier("This is a test to measure speed.")
end = time.time()

print(f"Tiempo de inferencia: {end - start:.4f} segundos")
```
